```python
# add modules to Python's search path
import os, sys
from pathlib import Path
script_dir = Path(os.path.abspath(''))
module_dir = str(script_dir.parent)
print(script_dir)
print(module_dir)
sys.path.insert(0, module_dir + '/modules')

# import rest of the modules
import numpy as np
import tensorflow as tf
import solver as sol
import time
```

```python
sigma = np.sqrt(2.)

def mu(x):
    x0, x1 = tf.split(x, [1, 1], axis=1)
    z = 4.0 * tf.reduce_sum(x0**2 + x1**2 - 1.0, axis=1)
    return tf.concat([x0*z, x1*z], axis=1)


def f(t, x, y, z):
    x0, x1 = tf.split(x, [1, 1], axis=1) 
    z_ = 4.0 * tf.reduce_sum(x0**2 + x1**2 - 1.0, axis=1)
    return - 4. * (z_+2.) * y

def g(x):
    x0, x1 = tf.split(x, [1, 1], axis=1)
    r = tf.reduce_sum(x0**2 + x1**2 - 1.0, axis=1)
    return tf.math.exp(-r) / (2.0 * np.pi)

solver = sol.DeepBSDE(mu=mu, fun_f=f, fun_g=g, dir='../data', name='test_solver')
```

```python
epochs = 40
solver.train(epochs)
```
